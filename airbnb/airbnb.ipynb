```python


fig, axs = plt.subplots(3, 1, figsize=(16, 12))
axs[0].bar(listings.columns, unique_count(listings))
axs[1].scatter(listings.columns, unique_count(listings))
axs[2].plot(listings.columns, unique_count(listings))
fig.suptitle('Categorical Plotting')
```

```python
fig, axs = plt.subplots(1, 1, figsize=(16, 12))
y_pos = np.arange(len(listings.columns))
 
# Create horizontal bars
plt.barh(y_pos, unique_count(listings))
 
# Create names on the y-axis
plt.yticks(y_pos, listings.columns)
plt.title("Percentage of unique attributes per field for Barcelona listings", fontsize=20)
# Show graphic
plt.show()
```
